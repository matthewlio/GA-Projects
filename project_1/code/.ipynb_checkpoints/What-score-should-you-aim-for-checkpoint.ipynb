## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

High school students often know which colleges and field of studies they would like to consider, but rarely know what SAT or ACT score they should aim for when applying to these colleges and majors. In this analysis, we explore the schools that have the highest and lowest SAT and ACT score requirements and see if there is a relationship between college prestige and test scores. Average scores of the test takers that applied for each college major will also be analysed, and insights will be gathered and concluded with recommendations for ideal scores to aim for, for highest success chance for the enrolment into students' ideal college and major.

### What Makes a School Prestigious?

Prestigious colleges. They are like brand names, everyone knows them, and your credibility and potential as a successful person would fly right up the charts as a student studying in one of them. You get better opportunities, higher chances of getting into top universities and jobs. It is not easy though; prestigious colleges are highly competitive to get in. #### Reputation

Reputation is an important aspect of being prestigious. To gain reputation, typically a college would need to consistently produce top graduates every year. Money also needs to be invested into top lecturers from around the country, as well as having proper facilities and laboratories. Reputation is then established by climbing up the school rankings, fairing phenomenal scores through surveys conducted by school administrators, college presidents and deans.

Students from top schools generally have very impressive SAT or ACT scores, and are usually the top of their high school classes. Later in our dataset, selectivity will translate to numbers through high admission score requirements, high number of applicants, and low acceptance rates.

Having money helps. With large fundings, schools would have the ability to attract the best researches, professors and lecturers from around the world. Good facilities, research, and laboratories can also be built, and resources can also be spent on its students, or to create a more convenient and conducive environment. Prestigious schools have the ability to acquire huge funding, and they invest heavily on improving themselves to become better schools consistently.

#### Faculty

Elite schools are able to attract professors, researchers and lecturers who are at the top of their fields. These faculties then generate more impressive results, like top students and research, thus driving up the school's ability to earn more funding as well as school prestige.

## Library Imports and Coding Challenges

1. Mean:

```python
# Coding Challenge 1: Manually calculate mean
# Function to calculate mean of a list
def get_mean(ls):
    total = 0
    for i in ls:
        total += i
    return total / len(ls)

# Test
x = [124, 523, 643, 532, 778, 868, 245]
get_mean(x)
```

2. Standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

```python
# Coding Challenge 2: Manually calculate standard deviation
def get_sd(ls):
    total = 0
    for i in ls:
        total += ((i - get_mean(ls))**2)
    return (total / len(ls))**(1/2)

# Test
x = [4, 9, 11, 12, 17, 5, 8, 12, 14]
get_sd(x)
```

3. Data cleaning function

```python
# Coding Challenge 3: Data cleaning function
# Function to convert percentage numbers of a list (string/object type) to decimals (float)
def convert_pct(ls):
    new_ls = []
    for i in ls:
        new_ls.append(float(i.split('%')[0]) / 100)
    return new_ls

# Test
x = ['1%', '25.7%', '37.8%', '48%', '55.5%', '69%', '88.8%', '90%']
convert_pct(x)
```

## Data Import and Cleaning

### Dataset 1: Schools' admittance scores and acceptance rates

```python
# First dataset
df_college = pd.read_csv('../data/sat_act_by_college.csv')
```

```python
df_college.head(1)
```

```python
# Dropping unnecessary columns
df_college = df_college.drop(columns = ['Test Optional?',
                                        'Applies to Class Year(s)',
                                        'Policy Details'], axis = 1).reset_index(drop=True)
```

```python
# Splitting the 25th and 75th percentile scores
# And adding median (50th percentile) column
# Assuming the 50th percentile lies directly between 25th and 75th percentile

# SAT 25th percentile
df_college['sat_25th_perc'] = pd.to_numeric(df_college['SAT Total 25th-75th Percentile'].apply(lambda x: x.split('-')[0])) 

# SAT 75th percentile
df_college['sat_75th_perc'] = pd.to_numeric(df_college['SAT Total 25th-75th Percentile'].apply(lambda x: x.split('-')[-1]))

# SAT median (50th percentile)
df_college['sat_median'] = (df_college['sat_25th_perc'] + df_college['sat_75th_perc']) / 2

# SAT median score, as a percentage
# Note: Subtraction of 400 as the minimum score for SAT is 400
# i.e. A score of 400 would be at 0%
df_college['sat_median_pct'] = ((df_college['sat_median'] - 400) / (1600 - 400) * 100).round(2)

# ACT 25th percentile
df_college['act_25th_perc'] = pd.to_numeric(df_college['ACT Total 25th-75th Percentile'].apply(lambda x: x.split('-')[0]))

# ACT 75th percentile
df_college['act_75th_perc'] = pd.to_numeric(df_college['ACT Total 25th-75th Percentile'].apply(lambda x: x.split('-')[-1]))

# ACT median (50th percentile)
df_college['act_median'] = (df_college['act_25th_perc'] + df_college['act_75th_perc']) / 2

# ACT median score, as a percentage
df_college['act_median_pct'] = (df_college['act_median'] / 36 * 100).round(2)
```

```python
# Checking for missing values
df_college.isnull().sum()
```

```python
# Deleting missing values
df_college.dropna(inplace = True)
df_college = df_college.reset_index(drop = True)
```

```python
# Displaying data types for each column
df_college.dtypes
```

```python
# Fixing incorrect data types
# Using previously defined function to convert 'Accept Rate' object into percentage
df_college['Accept Rate'] = convert_pct(df_college['Accept Rate'])
```

```python
# Dropping unnecessary columns
df_college = df_college.drop(columns = ['SAT Total 25th-75th Percentile',
                                        'ACT Total 25th-75th Percentile'],axis = 1).reset_index(drop=True)
```

```python
# Renaming columns
new_columns_dict = {
    'Accept Rate': 'accept_rate',
    'School': 'school',
    'Number of Applicants': 'no_of_applicants'
}
df_college.rename(columns = new_columns_dict, inplace = True)
```

```python
# Adding column for number of accepted applicants
df_college['applicants_accepted'] = (df_college['no_of_applicants'] * df_college['accept_rate']).astype(int)
```

```python
df_college.head()
```

```python
# Second dataset
df_major = pd.read_csv('../data/sat_2019_by_intended_college_major.csv')
```

```python
df_major.head(1)
```

```python
# Checking for missing values
df_major.isnull().sum()
```

```python
# Displaying data types for each column
df_major.dtypes
```

```python
# Fixing incorrect data types
# Converting 'TestTakers' to integer
df_major['TestTakers'] = pd.to_numeric(df_major['TestTakers'].apply(lambda x: x.replace(',', '')))
```

```python
# Using previously defined function to convert to percent float
df_major['Percent'] = convert_pct(df_major['Percent'])
```

```python
# Renaming columns
new_columns_dict = {
    'IntendedCollegeMajor': 'int_college_major',
    'TestTakers': 'test_takers',
    'Percent': 'test_takers_pct',
    'Total': 'total_score',
    'ReadingWriting': 'read_write',
    'Math': 'math'
}
df_major.rename(columns = new_columns_dict, inplace = True)
```

```python
# Creating new columns
# SAT mean score, as a percentage
# Note: Subtraction of 400 to the total score as the minimum score for SAT is 400
# i.e. A score of 400 would be at 0%
df_major['total_score_pct'] = ((df_major['total_score'] - 400) / (1600 - 400) * 100).round(2)

# Converting each section's score into a percentage score of its own section
# Note: Subtraction of 200 to each section's score as the minimum score for each section is 200
df_major['read_write_pct'] = ((df_major['read_write'] - 200) / (800 - 200) * 100).round(2)
df_major['math_pct'] = ((df_major['math'] - 200) / (800 - 200) * 100).round(2)
```

```python
def heatmap(data, title):
    plt.figure(figsize = (16,9))
    plt.title(title)

    # Get correlation of variables.
    corr = data.corr()

    # Set up mask to be "True" in the upper triangle.
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    # Plot our correlation heatmap, while masking the upper triangle to be white.
    with sns.axes_style("white"):
        sns.heatmap(corr, mask = mask, square = True, annot = True, vmin = -1, vmax = 1, linewidths = .5)
```

```python
# Function for plotting histogram subplots

def hist_subplot(data, title, xname, yname):
    if 'ACT' in title or 'R&W' in title:
        plt.figure(figsize=(15,5))
        plt.subplot(1,2,1)
        plt.hist(data, bins = 20, color = u'#ff9f1e')
        
    elif 'Math' in title:
        plt.figure(figsize=(15,5))
        plt.subplot(1,2,1)
        plt.hist(data, bins = 20, color = u'#5ca03c')
        
    else:
        plt.figure(figsize=(15,5))
        plt.subplot(1,2,1)
        plt.hist(data, bins = 20)
    
    # Median line
    plt.axvline(data.mean(), color='r', linestyle='dashed', linewidth=3)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.02, max_ylim*0.8, 'Mean: {:.0f}'.format(data.mean()))
    
    # Labels
    plt.grid()
    plt.xlabel(xname, size = 12)
    plt.ylabel(yname, size = 12)
    plt.title(title, size = 15)
    
    # Cumulative
    if 'ACT' in title or 'R&W' in title:
        plt.subplot(1,2,2)
        plt.hist(data, bins = 20, cumulative = True, color = u'#ff9f1e')
    
    elif 'Math' in title:
        plt.subplot(1,2,2)
        plt.hist(data, bins = 20, cumulative = True, color = u'#5ca03c')
        
    else:
        plt.subplot(1,2,2)
        plt.hist(data, bins = 20, cumulative = True)
    
    # Median line
    plt.axvline(data.mean(), color='r', linestyle='dashed', linewidth=3)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.02, max_ylim*0.8, 'Mean: {:.0f}'.format(data.mean()))
    
    # Labels
    plt.grid()
    plt.xlabel(xname, size = 12)
    plt.ylabel(f'Cumulative {yname}', size = 12)
    plt.title(f'Cumulative {title}', size = 15)
```

#### Boxplots for schools' admittance scores

From the boxplots above, the range of ACT scores for admittance is higher than the range of SAT scores. That means in general, you would technically need to score **higher** in ACT as compared to SAT to enter the same school!

Does this mean it would be easier to admission into schools if you do SAT instead of ACT? Unfortunately, we cannot conclude that just from this information. What if the reason behind a higher ACT score requirement is due to ACT being the easier test?

The formula to convert SAT scores to percentages:

(SAT_score - 400) / (1600 - 400) * 100

I subtracted 400 from both the SAT score and the total score achievable (full score at 1600) before converting them as a percentage (the multiply by 100 part). The logic behind this is simple. The minimum score for SAT is 400. That is to say, a test taker will never get a score under 400; the test starts at 200 score per section (there are 2 sections), every correct answer adds a little score, and you are not penalised for any wrong answers. This basically tells us that the starting score, or "0" mark, starts at 400! It makes sense to subtract 400 to both the total test score and 1600 before converting to a percentage score.

#### Insights:
- In general, schools require a higher ACT score than a SAT score for admission

```python
def scat_plot(data_x, data_y, title, xname, yname):
    plt.figure(figsize = (7, 7))
    plt.scatter(data_x, data_y)
    
    # Labels
    plt.grid()
    plt.title(title, fontsize = 15)
    plt.xlabel(xname, fontsize = 12)
    plt.ylabel(yname, fontsize = 12)
    
def scat_subplot(data_x1, data_x2, data_y, title1, xname1, title2, xname2, yname):
    plt.figure(figsize = (15, 7))
    plt.subplot(1,2,1)
    plt.scatter(data_x1, data_y)
    plt.grid()
    plt.title(title1, fontsize = 15)
    plt.xlabel(xname1, fontsize = 12)
    plt.ylabel(yname, fontsize = 12)

    plt.subplot(1,2,2)
    plt.scatter(data_x2, data_y)
    plt.grid()
    plt.title(title2, fontsize = 15)
    plt.xlabel(xname2, fontsize = 12)
    plt.ylabel(yname, fontsize = 12)
```
